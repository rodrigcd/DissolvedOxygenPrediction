```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import tensorflow as tf
from tensorflow.python.framework import dtypes
from tensorflow.contrib import learn
```

```python
class Database(object):
    def __init__(self, n_samples, time_steps):
        self.batch_size = 50
        train_prop = 0.6
        val_prop = 0.2
        test_prop = 0.2
        x = np.linspace(0, 500, n_samples)
        y = 0.8*np.sin(x) + 1
        self.examples = list()
        self.labels = list()
        self.x_axis = list()
        #Getting examples
        for i in range(len(x) - time_steps - 1):
            self.examples.append(y[i:i+time_steps])
            #self.labels.append(y[i+1:i+1+time_steps])
            self.labels.append(y[i+time_steps])
            self.x_axis.append(x[i+time_steps])
        #to numpy
        self.examples = np.array(self.examples)
        self.labels = np.array(self.labels)
        self.x_axis = np.array(self.x_axis)
        #shuffle
        #self.random_index = np.random.permutation(self.examples.shape[0])
        #self.examples = self.examples[self.random_index,:]
        #self.labels = self.labels[self.random_index]
        #self.x_axis = self.x_axis[self.random_index]
        print(self.examples.shape)
        print(self.labels.shape)
        #Getting batches list
        n_batches = np.floor(self.examples.shape[0]/self.batch_size)
        self.batches = self.split_into_batches(self.examples)
        self.labels = self.split_into_batches(self.labels)
        self.x_axis = self.split_into_batches(self.x_axis)
        n_train = int(np.floor(n_batches*train_prop))
        n_val = int(np.floor(n_batches*val_prop))
        n_test = int(np.floor(n_batches*test_prop))
        print(n_train)
        self.train_batches = self.batches[:n_train]
        self.train_labels = self.labels[:n_train]
        self.train_x = self.x_axis[:n_train]
        self.val_batches = self.batches[n_train:(n_train+n_val)]
        self.val_labels = self.labels[n_train:(n_train+n_val)]
        self.val_x = self.x_axis[n_train:(n_train+n_val)]
        self.test_batches = self.batches[(n_train+n_val):(n_train+n_val+n_test)]
        self.test_labels = self.labels[(n_train+n_val):(n_train+n_val+n_test)]
        self.test_x = self.x_axis[(n_train+n_val):(n_train+n_val+n_test)]
        self.train_index = 0
        self.val_index = 0
        self.test_index = 0
        
    def get_train_batch(self):
        if self.train_index >= len(self.train_batches):
            self.train_index = 0
        self.train_index += 1
        return self.train_batches[self.train_index -1], self.train_labels[self.train_index - 1]
       
    def get_val_batch(self):
        if self.val_index >= len(self.val_batches):
            self.val_index = 0
        self.val_index += 1
        return self.val_batches[self.val_index -1], self.val_labels[self.val_index - 1]
    
    def get_test_batch(self):
        if self.test_index >= len(self.test_batches):
            self.test_index = 0
        self.test_index += 1
        return self.test_batches[self.test_index -1], self.test_labels[self.test_index - 1]

    def split_into_batches(self, input_array):
        batches = list()
        array_length = input_array.shape[0]
        index = 0
        while index <= array_length:
            batches.append(input_array[index:index+self.batch_size])
            index += self.batch_size
        return batches
```

```python
class lstm_cell(object):
    #TODO: DO THE SAME AS CONVNET
    def __init__(self, input_tensor, hidden_units, time_steps, layer_name):
        self.input_tensor = input_tensor
        self.time_steps = time_steps
        self.layer_name = layer_name
        self.hidden_units = hidden_units
        with tf.variable_scope(layer_name):
            self.lstm = tf.nn.rnn_cell.BasicLSTMCell(hidden_units)
            self.state = self.lstm.zero_state(input_tensor.get_shape()[0], dtype='float32')
            self.output_tensor, self.state = self.lstm(self.input_tensor, self.state)
            
    #def eval_lstm(self, input_tensor):
    #    state = self.lstm.zero_state(input_tensor.get_shape()[0], dtype='float32')
    #    output_tensor, state = self.lstm(input_tensor, state)
    #    return output_tensor, state
```

```python
class Model(object):
    
    def __init__(self):
        self.time_steps = 5
        self.iterations = 0
        self.database = Database(n_samples = 1000000, time_steps = self.time_steps)
        self.batch_size = self.database.batch_size
        self.sess = tf.Session()
        self.build_model()
        # ?????????????????
        self.init_op = tf.initialize_all_variables()
        self.sess.run(self.init_op)
        # ################
        self.optimizer = tf.train.GradientDescentOptimizer(5e-3)
        self.train_params = self.optimizer.minimize(self.loss)
        #self.correct_predictions = tf.equal(tf.argmax(self.model_output,1),
        #                                tf.argmax(self.target,1))
        self.accuracy = tf.reduce_mean(
            tf.square(self.model_output - self.target), name='accuracy')
    
    def build_model(self):
        self.model_input = tf.placeholder(tf.float32, shape = (self.batch_size, self.time_steps))
        self.target = tf.placeholder(tf.float32, shape = (self.batch_size,))
        
        self.lstm_1 = lstm_cell(self.model_input, hidden_units = 1, 
                                time_steps = self.time_steps, 
                                layer_name = 'lstm_1')
        self.model_output = self.lstm_1.output_tensor
        print("SHAPES")
        print(self.model_output.get_shape())
        print(self.target.get_shape())
        self.loss = tf.reduce_mean(tf.square(self.model_output - self.target),
                                  name = 'loss')
        #To eval model
        #self.eval_input = tf.placeholder(tf.float32)
        #self.output = self.lstm_1.eval_lstm(self.eval_input)
        
    def train_nbatch(self, n_iterations = 100, show_error = True):
        train_step = self.train_params
        self.iterations += n_iterations
        for iteration in range(n_iterations):
            data, labels = self.database.get_train_batch()
            val_data, val_labels = self.database.get_val_batch()
            _ = self.sess.run((train_step),feed_dict={
                self.model_input: data,
                self.target: labels,
                #self.keep_prob: 0.5
            })
            if iteration%10 == 0:
                val_accuracy = self.accuracy.eval(
                    feed_dict={
                        self.model_input: val_data,
                        self.target: val_labels,
                        #self.keep_prob: 0.5# to be fair
                        }, session=self.sess
                    )
                train_accuracy = self.loss.eval(
                    feed_dict={
                        self.model_input: data,
                        self.target: labels,
                        #self.keep_prob: 0.5# to be fair
                        }, session=self.sess
                    )
                if show_error:
                    print("val accuracy = "+ str(val_accuracy))
                    print("train accuracy = "+ str(train_accuracy))
        return val_accuracy, train_accuracy
    
    def evaluate(self):
        input_sequence = self.database.test_batches
        values = list()
        for batch in input_sequence:
            value = self.model_output.eval(
                feed_dict={
                    self.model_input: batch
                }, session = self.sess
            )
            values.append(value)
        return values
        
```

```python
test = lstm_model.evaluate()
test_x = lstm_model.database.test_x
test_labels = lstm_model.database.test_labels
test = np.concatenate(test, axis = 0)
test_x = np.concatenate(test_x, axis = 0)
test_labels = np.concatenate(test_labels, axis = 0)
plt.plot(test_x, test_labels)
plt.plot(test_x, test)
plt.show()
```
